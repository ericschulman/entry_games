```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.base.model import GenericLikelihoodModel
```

```python
entry_data = pd.read_csv("entry.csv",sep='\t')
entry_data = entry_data.dropna()
entry_data['HD_1'] = 1*(entry_data['HD'] > 0)
entry_data['LOW_1'] = 1*(entry_data['LOW'] > 0)
entry_data['income_2'] = np.log(1 + entry_data['income_per_capita'])
entry_data['population_2'] = np.log(1 + entry_data['population'])
entry_data = entry_data[ (entry_data['population']>=2000) & (entry_data['income_per_capita']>=15000) ]
print(entry_data)
```

```python
# Model fit for HD stores
x = entry_data[['income_2','population_2','LOW_1']].copy()
y = entry_data['HD_1']
model1 = sm.regression.linear_model.OLS(y,sm.add_constant(x)).fit()
model1.summary()
```

```python
# Model fit for HD stores
x = entry_data[['income_2','population_2','LOW']].copy()
y = entry_data['HD']
model1 = sm.regression.linear_model.OLS(y,sm.add_constant(x)).fit()
model1.summary()
```

```python
# Model fit for LO stores
x = entry_data[['income_2','population_2','HD_1']].copy()
y = entry_data['LOW_1']
model1 = sm.regression.linear_model.OLS(y,sm.add_constant(x)).fit()
model1.summary()
```

```python
# Model fit for LO stores
x = entry_data[['income_2','population_2','HD']].copy()
y = entry_data['LOW']
model1 = sm.regression.linear_model.OLS(y,sm.add_constant(x)).fit()
model1.summary()
```
